```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
x,y=make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
```

```python
#lets apply OLS
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red')
```

Assuming m=known=78.35
lets start with b=0

```python
y_pred=(78.35*x+0).reshape(4)

plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='OLS')
plt.plot(x,y_pred,color='#00a65a',label='b=0')

plt.legend()
plt.show()
```

Applying Gradient descent

```python
m=78.35
b=0
#ravel() functions returns contiguous flattened array(1D array with all the input-array elements and with the same type as it).
loss_slope=-2*np.sum(y-m*x.ravel()-b) 
loss_slope
```

```python
#lets take learning learning rate a 0.1

lr=0.1 #learning rate=lr

step_size=loss_slope*lr
step_size
```

```python
#calculating the new intercept
b=b-step_size
b
```

```python
y_pred1=(78.35*x+b).reshape(4)


plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='OLS')
plt.plot(x,y_pred,color='#00a65a',label='b=0')
plt.plot(x,y_pred1,color='black',label='b={}'.format(b))

plt.legend()
plt.show()
```

Iteration using epochs:

```python
b=-100
m=78.35
lr=0.1

epochs=10
for i in range(epochs):
    loss_slope=-2*np.sum(y-m*x.ravel()-b)
    b=b - (lr * loss_slope)

    y_pred= m * x + b

    plt.plot(x,y_pred,label=f'epoch= {i}')
plt.scatter(x,y)
plt.legend()
plt.show()
```
