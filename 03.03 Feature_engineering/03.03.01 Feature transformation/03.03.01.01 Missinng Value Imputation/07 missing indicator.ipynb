```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.impute import MissingIndicator,SimpleImputer
```

```python
df=pd.read_csv(r"A:\Work Docs\DATA Set\titanic_train.csv",usecols=['Age','Fare','Survived'])
```

```python
df.head()
```

```python
df.isnull().mean()*100
```

```python
x=df.drop(columns=['Survived'])
y=df['Survived']
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
x_train.head()
```

```python
si=SimpleImputer()
x_train_trf=si.fit_transform(x_train)
x_test_trf=si.transform(x_test)
```

```python
x_train_trf
```

```python
from sklearn.linear_model import LogisticRegression

clf=LogisticRegression()

clf.fit(x_train_trf,y_train)

y_pred=clf.predict(x_test_trf)

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)*100
```

# Process 1 (lengthy)
### import Missing class, Make object, Merge the column coming from object to original Data

```python
mi=MissingIndicator()
mi.fit(x_train)
```

```python
mi.features_
```

```python
x_train_missing=mi.transform(x_train)
x_test_missing=mi.transform(x_test)
x_train_missing
```

```python
x_test_missing
```

```python
x_train['Age_NA']=x_train_missing
x_train.sample(5)
```

```python
x_test['Age_NA']=x_test_missing
x_test.sample(5)
```

```python
x_train
```

```python
si=SimpleImputer()

x_train_trf2=si.fit_transform(x_train)
x_test_trf2=si.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

clf =LogisticRegression()

clf.fit(x_train_trf2,y_train)

y_pred=clf.predict(x_test_trf2)

from sklearn.metrics import accuracy_score

accuracy_score(y_test,y_pred)*100
```

# Precess 2 (Short and Clean)
### Just use-   add_indicator=True

```python
si=SimpleImputer(add_indicator=True)

x_train=si.fit_transform(x_train)
x_test=si.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

clf =LogisticRegression()

clf.fit(x_train,y_train)

y_pred=clf.predict(x_test)

from sklearn.metrics import accuracy_score

accuracy_score(y_test,y_pred)*100
```
