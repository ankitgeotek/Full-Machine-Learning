```python
import numpy as np 
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression, SGDRegressor

from sklearn.preprocessing import PolynomialFeatures, StandardScaler

from sklearn.metrics import r2_score

from sklearn.pipeline import Pipeline
```

```python
x= 6*np.random.rand(200,1)-3
y=0.8 * x**2 +0.9* x+ 2 + np.random.randn(200,1)

# y=0.8x^2+ 0.9x+2
```

```python
plt.plot(x, y,'b.')

plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
xtr, xtst, ytrn, ytst = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
lr=LinearRegression()
```

```python
lr.fit(xtr,ytrn)
```

```python
ypred=lr.predict(xtst)
```

```python
r2_score(ytst,ypred)
```

```python
plt.plot(x, y,'b.')
plt.plot(xtr,lr.predict(xtr),color='r')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#Applying  Polynomial Linear regression
# degree-2
poly=PolynomialFeatures(degree=2,include_bias=True)

xtr_trf=poly.fit_transform(xtr)
xtst_trf=poly.transform(xtst)
```

```python
print(xtr_trf[0])
print(xtst_trf[0])
```

```python
lr=LinearRegression()

lr.fit(xtr_trf,ytrn)
```

```python
ypred=lr.predict(xtst_trf)
```

```python
r2_score(ytst,ypred)
```

```python

plt.plot(xtr_trf,lr.predict(xtr_trf),color='r')

plt.plot(x, y,'b.')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
