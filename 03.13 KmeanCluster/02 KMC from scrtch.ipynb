```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

from kmeans import KMeans


```

```python
centroids = [(-10,10), (-10,-10), (-5,5), (-5,-5)]
cluster_std = [1, 1, 1, 1]
```

```python
x, y = make_blobs(n_samples=100, cluster_std = cluster_std, centers= centroids, n_features= 2, random_state= 2)
```

```python
plt.scatter(x[:,0], x[:,1])
```

```python
kms = KMeans(n_clusters= 4, max_itteration= 1000)
```

```python
y_mean = kms.fit_predict(x)
```

```python
y_mean
```

```python
plt.scatter(x[y_mean == 0,0], x[y_mean == 0,1], color= 'red')
plt.scatter(x[y_mean == 1,0], x[y_mean == 1,1], color= 'blue')
plt.scatter(x[y_mean == 2,0], x[y_mean == 2,1], color= 'green')
plt.scatter(x[y_mean == 3,0], x[y_mean == 3,1], color= 'black')
```
